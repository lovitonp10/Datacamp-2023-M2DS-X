```python
def dataset_maker(labels, data):
    
    # Processing of training labels
    labels.begin = pd.to_datetime(labels.begin)
    labels.end = pd.to_datetime(labels.end)
    labels = labels.apply(lambda x: pd.date_range(start=x['begin'], end=x['end'], freq='10T'), axis=1)
    labels = labels.explode()
    labels = labels.reset_index().drop('index',axis=1)
    
    # Processing of training dataset
    ## Créer un nouvel index comprenant toutes les dates manquantes
    date_range = pd.date_range(start=data.index.min(), end=data.index.max(), freq='10T')
    data = data.reindex(date_range)#, method='ffill')
    ## Remplir les valeurs manquantes
    data.interpolate(inplace=True)
    
    # Creation of target feature
    data['target'] = data.index.isin(labels[0]).astype(int)
    
    return data
```

```python
data_train_final = dataset_maker(labels_train, data_train)
data_test_final = dataset_maker(labels_test, data_test)
```

```python
data_train_final.to_csv("data_final/data_train.csv")
data_test_final.to_csv("data_final/data_test.csv")
```
